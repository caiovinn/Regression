#**Ajustar um Modelo de Regressão Linear Múltipla**

Ajuste um modelo de regressão linear múltipla considerando todas as variáveis do seu conjunto de
dados. Apresente os valores numéricos da matriz $X^{T}X$ e sua inversa. Assim como da matriz $X^TY$
e estimativas do vetor dos $\beta$'s e interprete. Explique se existe diferença entre o EMV e o estimador de
mínimos quadrados de $\beta$? justifique.


Para o modelo de Regressão Linear Múltipla:


\begin{align}
Y = \beta_0 + \beta_1 X_1 + .... \beta_k X_k + \epsilon 
\end{align}

Temos como  estimador pelos minimos quadrados:

\begin{align}
\widehat{Y} = X\widehat{\beta} = X(X^{T}X)^{-1}X^{T} Y
\end{align}




```python
#Adicionando a constante no X para o Beta0
x = sm.add_constant(x)
x_transposta = x.T
x_transposta_x = x_transposta.dot(x)
inversa_x_transposta_x = np.linalg.inv(x_transposta_x)
x_transposta_y = x_transposta.dot(y)
```

Matriz $X^TX$

```python
round(x_transposta_x,2)
```

Matriz $(X^TX)^{-1}$

```python
print(inversa_x_transposta_x.round(2))
```

Matriz $X^TY$

```python
x_transposta_y
```

Cálculo dos Coeficientes 

```python
coeficientes = inversa_x_transposta_x.dot(x_transposta_y)
for i in range(len(coeficientes)):
  print(f'Beta{i} = {coeficientes[i]}')

beta0 = coeficientes[0]
beta1 = coeficientes[1]
beta2 = coeficientes[2]
beta3 = coeficientes[3]
beta4 = coeficientes[4]
```

Temos então para o Modelo de Regressão Linear Múltipla
\begin{align}
\widehat{Y} = 5.8945 -0.4779 x_1 + 0.1827 x_2 + 35.4028x_3 + 5.8439x_4
\end{align}
